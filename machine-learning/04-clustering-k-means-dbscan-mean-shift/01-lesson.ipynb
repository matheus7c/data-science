# Normalizando os dados do DataFrame

## Normalização manual

```python
atributos = df.columns
```

```python
atributos
```

```python
df_normalizado = df
```

```python
atributos = df_normalizado.columns
```

```python
df_normalizado
```

```python
for atributo in atributos:
    df_normalizado[atributo] = (df_normalizado[atributo]-min(df_normalizado[atributo]))/(max(df_normalizado[atributo])-min(df_normalizado[atributo]))
```

```python
df_normalizado
```
